Que.2 How does broadcasting work in NumPy?

   -> Broadcasting in NumPy allows you to perform arithmetic operations on arrays of different shapes in a way that avoids making unnecessary copies of data. It’s a powerful concept that lets NumPy "stretch" the smaller array along dimensions so that shapes are compatible for element-wise operations.

Basic Broadcasting Rules
Two arrays are compatible for broadcasting if, for each dimension, one of the following is true:

The dimensions are equal, or

One of the dimensions is 1

If these conditions are met, NumPy will broadcast the smaller array across the larger one so that they have the same shape.

NumPy compares shapes from right to left. When the shapes don't match in length, the shorter one is padded on the left with ones.

Que.6 What are the differences between NumPy arrays and Python lists?

  ->  NumPy arrays and Python lists are used to store collections of data, they have some key differences in terms of performance, functionality, and behavior. Here's a comparison:

  1. Data Type Consistency
NumPy arrays: Homogeneous — all elements must be of the same data type (e.g., all floats or all integers).

Python lists: Heterogeneous — elements can be of different types (e.g., int, float, str).

2. Performance
NumPy arrays: Much faster for numerical operations due to optimized C backend and vectorized operations.

Python lists: Slower because each element is a separate Python object with overhead.

3. Functionality
NumPy arrays: Support advanced mathematical operations like matrix multiplication, broadcasting, Fourier transforms, linear algebra, etc.

Python lists: Limited to basic operations (e.g., append, pop, slicing).

4. Memory Efficiency
NumPy arrays: Use less memory by storing elements in a contiguous block with fixed-size data types.

Python lists: More memory overhead since each element is a full-fledged Python object.

5. Vectorized Operations
NumPy arrays: Support element-wise operations directly (e.g., a + b adds arrays element-wise).

Python lists: Require loops or list comprehensions (e.g., [x + y for x, y in zip(a, b)]).

6. Indexing and Slicing
NumPy arrays: More powerful indexing, including boolean masking and fancy indexing.

Python lists: Basic indexing and slicing.

Que.21 Why is NumPy’s array slicing faster than Python’s list slicing?

   ->  NumPy’s array slicing is faster than Python’s list slicing due to
       several key differences in how NumPy and Python handle data and memory:

1. Contiguous Memory Allocation (NumPy):
NumPy arrays are implemented in a way that the data is stored in contiguous memory blocks (in C-style arrays). This allows for direct access to memory locations and efficient manipulation of the data.

When you slice a NumPy array, it doesn't create a copy of the data. Instead, it returns a view on the original array, meaning it points to the same memory location but with adjusted indexing. This allows NumPy to avoid extra memory allocations and speed up slicing.

2. Dynamic Array (Python Lists):
Python lists are more general-purpose containers that can hold elements of any type. Internally, Python lists are implemented as arrays of pointers to objects, not the objects themselves. This means that slicing a list requires more overhead since Python needs to handle each element individually and might need to allocate new memory to store the sliced portion.

When you slice a Python list, a new list object is created, and it copies the relevant items into this new list. This copying operation adds extra time.

3. Optimized Low-Level Operations (NumPy):
NumPy is built using highly optimized C code under the hood. When you perform operations on NumPy arrays, it leverages low-level, vectorized operations that are more efficient than Python’s built-in list operations. These operations are compiled and CPU optimized, making slicing much faster in NumPy than Python lists.

NumPy's slicing also utilizes advanced indexing techniques that are designed specifically for handling large arrays efficiently.

4. Python's List Slicing Creates a New List:
Python lists create a new list object when sliced, which involves copying the elements from the original list into a new list. This copying process consumes both time and memory.

NumPy slices, on the other hand, create views rather than copies. This is why slicing a NumPy array does not involve any data copying, which leads to significant performance improvements when working with large datasets.

5. Memory Efficiency (NumPy):
NumPy is more memory-efficient than Python lists, especially for large datasets. This is because NumPy arrays are homogeneous (all elements are of the same type and size), whereas Python lists are heterogeneous and require additional overhead for storing type information for each element.

As a result, NumPy arrays use less memory for storing the data, which contributes to their faster slicing performance.

Summary of the Key Differences:
Memory Layout: NumPy arrays are contiguous in memory, while Python lists store references to objects.

Memory Allocation: NumPy slicing uses views (no copy), while Python list slicing creates new copies.

Low-Level Optimization: NumPy is implemented in highly optimized C code, whereas Python lists are more flexible and slower for specific operations like slicing.

Data Homogeneity: NumPy arrays store elements of the same type and size, making memory access faster compared to Python’s heterogeneous lists.

```python
#Que.4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example DataFrame with random data
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [5, 6, 7, 8, 9]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
#Que.5 Generate a bar plot using Plotly.

import plotly.graph_objects as go

# Data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [10, 15, 7, 20]

# Create a bar plot
fig = go.Figure(data=[go.Bar(
    x=categories,  # X-axis categories
    y=values,      # Y-axis values
    marker=dict(color='blue')  # Bar color
)])

# Add title and labels
fig.update_layout(
    title='Fruit Bar Plot',
    xaxis_title='Fruit',
    yaxis_title='Quantity',
)

# Show the plot
fig.show()
```

```python
#Que.7 Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

# Define two arrays
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# Element-wise multiplication
result = a * b

# Display the result
print("Array a:", a)
print("Array b:", b)
print("Element-wise multiplication:", result)
```

```python
#Que.8 import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]
y3 = [3, 5, 2, 6, 9]

# Create line plot
plt.plot(x, y1, label='Line 1', color='blue', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--', marker='s')
plt.plot(x, y3, label='Line 3', color='red', linestyle='-.', marker='^')

# Add title and labels
plt.title('Line Plot with Multiple Lines')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show legend
plt.legend()

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
#Que.9 Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd
import numpy as np

# Create sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [85, 92, 78, 90, 88]
}

df = pd.DataFrame(data)

# Set threshold
threshold = 85

# Filter rows where 'Score' > threshold
filtered_df = df[df['Score'] > threshold]

# Display results
print("Original DataFrame:")
print(df)
print("\nFiltered DataFrame (Score > {}):".format(threshold))
print(filtered_df)
```

```python
#Que.10 Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Create histogram
sns.histplot(data, bins=30, kde=True, color='skyblue')

# Add titles and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
#Que.11 Perform matrix multiplication using NumPy.

import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication
C = np.dot(A, B)

# Display result
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("A * B =\n", C)
```

```python
#Que.13 Create a 3D scatter plot using Plotly.

import plotly.graph_objs as go
import plotly.io as pio
import numpy as np

# Sample data
np.random.seed(42)
x = np.random.randn(50)
y = np.random.randn(50)
z = np.random.randn(50)

# Create the 3D scatter plot
scatter = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=6,
        color=z,                # Color by z-axis values
        colorscale='Viridis',  # Color scale
        opacity=0.8
    )
)

layout = go.Layout(
    title='3D Scatter Plot Example',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

fig = go.Figure(data=[scatter], layout=layout)

# Show the plot
pio.show(fig)
```
